```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots()

# Create a circle with center (0.5, 0.5) and radius 0.3
circle = patches.Circle((0, 0), 0.3, edgecolor='blue', facecolor='none', linewidth=2)

# Add the circle to the axes
ax.add_patch(circle)

# Add horizontal and vertical lines passing through the center of the ellipse
ax.axhline(y=0.5, color='red', linestyle='--', linewidth=1)  # Horizontal line at y=0.5
ax.axvline(x=0.5, color='blue', linestyle='--', linewidth=1)  # Vertical line at
# Set the limits of the plot
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_aspect('equal')  # Ensures the circle is not skewed

# Show the plot
plt.show()
```

```python
import sympy as sp
import numpy as np

# Define the variable z
z = sp.symbols('z')

# Define the function f
f = 1 / ((z - 1) * (z - 2))

# Define the radius of the contour
radius = 1.5

# Parameterization of the contour |z| = 1.5
theta = sp.symbols('theta', real=True)  # Angle parameter
z_contour = radius * sp.exp(sp.I * theta)

# Compute the integral over the contour
# dz = r * i * exp(i * theta) dtheta
dz = sp.diff(z_contour, theta)

# Substitute the contour parameterization into the function
f_substituted = f.subs(z, z_contour)

# Compute the integral
integral = sp.integrate(f_substituted * dz, (theta, 0, 2 * sp.pi))

# Display the result
integral
```

```python
import sympy as sp

# Define the variable theta as the parameter for the circle
theta = sp.symbols('theta', real=True)

# Parameterize the contour |z - 3i| = 2
z = 2 * sp.exp(sp.I * theta) + 3 * sp.I  # Circle centered at z=3i with radius 2

# Define the function f
f = 1 / z


# Find the singularities (non-analytic points)
singularities = sp.singularities(f, z)
# Compute the differential dz
dz = sp.diff(z, theta)

# Compute the integral over the contour
integral = sp.integrate(f * dz, (theta, 0, 2 * sp.pi))

# Display the result
print(singularities)
print(integral)
```
